```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('mobile-use.csv')
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
sns.scatterplot(df, x = 'Age', y = 'spComfort', alpha = .2)
plt.show
```

```python
sns.scatterplot(df, x = 'Age', y = 'spComfort', hue = 'Ethnicity', alpha = .2)
plt.show
```

```python
g = sns.FacetGrid(df, col = 'Ethnicity', col_wrap = 3)
g.map(sns.histplot, 'spComfort')
plt.show()
```

```python
sns.scatterplot(df[df['Age'] > 65], x = 'Age', y = 'spComfort', alpha = .2)
plt.show()
```

```python
sns.stripplot(df, x = 'spBrand', y = 'spComfort')
plt.show()
```

```python
sns.boxplot(df, x = 'spBrand', y = 'spComfort')
```

```python
sns.lineplot(df, x = 'Age', y = 'spComfort')
```

```python
sns.regplot(data = df, x = 'Age', y = 'spComfort', y_jitter = .05)
plt.show()
```

```python
sns.boxplot(data = df, x = 'spBrand', y = 'spComfort')
```

```python
sns.scatterplot(df, x = 'spBrand', y = 'Age', alpha = .2)
```
