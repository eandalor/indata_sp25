```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('insurance.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
sns.relplot(data = df, x = 'bmi', y = 'charges', col = 'sex', kind = 'scatter')
plt.show
```

```python
sns.relplot(data = df, x = 'bmi', y = 'charges', col = 'sex', kind = 'scatter', size = 'region')
plt.show
```

```python
df['smoker_id'] = df['smoker'].map({'yes': 'Smoker', 'no': 'Non-smoker'})
```

```python
df.head()
```

```python
sns.relplot(data = df, x = 'bmi', y = 'charges', col = 'sex', kind = 'scatter', size = 'region', hue = 'smoker_id', alpha = 0.5)
plt.show
```
