```python
import pandas as pd
import plotly.graph_objects as go

#from ipywidgets import widgets
#from dash import Dash, dcc, html
```

```python
meteor = pd.read_csv('Meteorite_Landings_Sample.csv')
meteor.head()
```

```python
meteor.index
```

```python
# Create a scatterplot map of meteorite landings
fig1 = go.Figure(data=go.Scattergeo(
        lon = meteor['reclong'],
        lat = meteor['reclat'],
        mode = 'markers',
        ))

fig1.update_layout(
        title = 'Meteorite Landings',
        geo_scope = 'world',
    )

fig1.show()
```

```python
# Change color and symbol
fig2 = go.Figure(data=go.Scattergeo(
        lon = meteor['reclong'],
        lat = meteor['reclat'],
        text = meteor['mass (g)'],
        mode = 'markers',
        marker = dict(
            size = 8,
            opacity = 0.8,
            symbol = 'star-diamond',
            color = 'red'
        )
        ))

fig2.update_geos(projection_type = 'equirectangular')

fig2.update_layout(
        title = 'Meteorite Landings',
        geo_scope = 'world',
    )

fig2.show()
```

```python
# Map color to mass, color map
fig3 = go.Figure(data=go.Scattergeo(
        lon = meteor['reclong'],
        lat = meteor['reclat'],
        text = meteor['mass (g)'],
        mode = 'markers',
        marker = dict(
            size = 6,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'circle',
            colorscale = 'electric',
            color = meteor['mass (g)'],
            colorbar = dict(
                title=dict(
                    text = 'Meteorite Mass (g)'
                ))
        )
        ))

fig3.update_geos(
    showland = True, landcolor = 'lightgrey',
    showocean = True, oceancolor = 'lightblue',
    showlakes = True, lakecolor = 'grey',
    showrivers = True, rivercolor = 'grey',
    projection_type = 'equirectangular',
)

fig3.update_layout(
        title = 'Meteorite Landings',
        geo_scope = 'world',
    )

fig3.show()
```

```python
# Globe

fig4 = go.Figure(data=go.Scattergeo(
        lon = meteor['reclong'],
        lat = meteor['reclat'],
        text = meteor['mass (g)'],
        mode = 'markers',
        marker = dict(
            size = 6,
            opacity = 0.8,
            reversescale = True,
            autocolorscale = False,
            symbol = 'circle',
            colorscale = 'electric',
            color = meteor['mass (g)'],
            colorbar = dict(
                title=dict(
                    text = 'Meteorite Mass (g)'
                ))
        )
        ))

fig4.update_geos(
    showland = True, landcolor = 'lightgrey',
    showocean = True, oceancolor = 'lightblue',
    showlakes = True, lakecolor = 'grey',
    showrivers = True, rivercolor = 'grey',
    projection_type = 'orthographic',
)

fig4.update_layout(
        title = 'Meteorite Landings',
        geo_scope = 'world',
    )

fig4.show()
```

```python
#Dropdown, map, globe
fig5 = go.Figure()

fig5.add_trace(
    go.Scattermap(
        lon = meteor['reclong'],
        lat = meteor['reclat'],
        text = meteor['mass (g)'],
        mode = 'markers',
        marker = dict(size = 8, symbol = 'marker')
        ))

fig5.update_geos(
    showland = True, landcolor = 'lightgrey',
    showocean = True, oceancolor = 'lightblue',
    showlakes = True, lakecolor = 'grey',
    showrivers = True, rivercolor = 'grey',
    projection_type = 'orthographic',
)

fig5.update_layout(
        title = 'Meteorite Landings',
        geo_scope = 'world',
        updatemenus = [
        dict(
            buttons = list([
                        dict(
                    args = ["type", "scattermap"],
                    label = "Map",
                    method = "restyle"
                ),
                dict(
                    args = ["type", "scattergeo"],
                    label = "Globe",
                    method = "restyle"
                ),
            ]),
            direction="down",
            pad={"r": 10, "t": 10},
            showactive=True,
            x=0.1,
            xanchor="left",
            y=1.1,
            yanchor="top"
        ),
    ]
    )

fig5.show()
```

```python
app = Dash()
app.layout = html.Div([
    dcc.Graph(figure = fig5)
])

app.run_server(debug=False, use_reloader=False)
```
